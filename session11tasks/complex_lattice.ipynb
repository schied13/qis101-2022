```python
# Cell 1 complex_lattice
#solution provided by Dave Biersach


import matplotlib.pyplot as plt
import numpy as np 
from matplotlib.ticker import MultipleLocator


%matplotlib widget

#function that we are plotting
def f(z):
    return z**2 +z*1j +1

plt.close("all")
fig = plt.figure(" ")
ax = fig.add_subplot()

# real values
x = np.linspace(-20,20,1000)
# imaginary values
y = np.linspace(0,20,1000)

#meshgrid of x and y matching each valueto one another
x,y = np.meshgrid(x,y)

# call function that we are mapping
z= f(x+y*1j)

#plot real part of z on the x and imaginary on y
ax.scatter(np.real(z), np.imag(z),s=(72.0/plt.gcf().dpi)**2) 

ax.set_title("f(z)")
ax.set_xlabel("Re")
ax.set_ylabel("Im")

ax.grid()
ax.set_xlim(-11,11)
ax.set_ylim(-11,11)

ax.set_aspect("equal")

plt.show()
```

```python
#Cell 2 lattice points

plt.close("all")
fig = plt.figure(" ")
ax = fig.add_subplot()

# declare bounds of f(z)
x = np.linspace(-10,10,21)
y = np.linspace(-10,10,21)

# match x and y values
x,y = np.meshgrid(x,y)


#calculate which points are lattice
x,y = np.where(x>=y**2+1,x,None),np.where(x>=y**2 +1,y, None)

#determine number of points from x pos
num_pts = np.sum((x.flatten() != None))
#display on scatter plot 
ax.scatter(x,y,zorder = 2)


print(f"Total nuber of lattice points inside curve = {num_pts:,}")


ax.set_title("Lattice Points")
ax.set_xlabel("Re")
ax.set_ylabel("Im")

ax.grid()

ax.set_xlim(-11,11)
ax.set_ylim(-11,11)

ax.set_aspect("equal")

plt.show()
```
