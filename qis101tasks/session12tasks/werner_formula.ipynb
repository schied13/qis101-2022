```python
# werner_formula.ipynb
# Cell 1


import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import AutoMinorLocator


def plot(ax):
    #frequency of each sine wave
    a = 0.8
    b = 0.5
    
    #calculating f_1(x)= sin
    x0 = np.linspace(-3 * np.pi, 3 * np.pi, 100)
    y0 = np.sin(a * x0)

    #calcualting f_2(x)
    x1 = np.linspace(-3 * np.pi, 3 * np.pi, 100)
    y1 = np.sin(b * x1)
    
    #Calculating f_3(x)
    x2 = np.linspace(-3 * np.pi, 3 * np.pi, 100)
    y2 = np.sin(a * x2) * np.sin(b * x2)

    #Calculating f_(x) using werners formulas
    x3 = np.linspace(-3 * np.pi, 3 * np.pi, 100)
    y3 = 0.5 * (np.cos(a * x3 - b * x3) - np.cos(a * x3 + b * x3))

    ax.set_title(r"Werner's Formula")
    ax.set_xlabel("x")
    ax.set_ylabel("y")

    ax.axhline(y=0.0, color="lightgray")
    
    ax.plot(x0, y0, label=r"$f_1(x) = \sin(0.8x)$")
    ax.plot(x1, y1, label=r"$f_2(x) = \sin(0.5x)$")
    ax.plot(x2, y2, label=r"$f_3(x) = f_1(x)*f_2(x)$")
    ax.plot(x3, y3, marker = "o", markerfacecolor = "none",markersize = 4, linestyle = "dotted",label=r"$f_4(x) = $Werner Formula for $ f_1(x)*f_2(x)$")
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator())


def main():

    plt.close("all")
    fig = plt.figure(" ")
    gs = fig.add_gridspec(1, 1)

    ax = fig.add_subplot(gs[0, 0])
    plot(ax)
    plt.legend(fontsize="small")
    plt.show()


main()
```
