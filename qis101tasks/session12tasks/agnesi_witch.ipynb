```python
#agnesi_witch.ipynb
# Cell 1

%matplotlib widget

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import AutoMinorLocator

def ag_witch(num_terms,x):
    sum0 = 0
    for n in range(0,num_terms+1):
        sum0 +=np.power(-1,n)*np.power(x,2*n)
    return sum0

def plot(ax):
    a = -1.5
    b = 1.5
    asymptotey = np.linspace(-2,2,100)
    asymptotex= np.ones(100)
    asymptotey2 = np.linspace(-2,2,100)
    asymptotex2= np.ones(100)*-1
    x = np.linspace(a, b,100)
    y = 1/(x**2+1)
    
    x2 = np.linspace(a, b,100)
    y2= ag_witch(2,x2)
    
    x3 = np.linspace(a, b,100)
    y3 =ag_witch(3,x3)
    
    x4 =np.linspace(a, b,100)
    y4 =ag_witch(4,x4)
    
    x5 =np.linspace(a, b,100)
    y5 =ag_witch(5,x5)
    
    x6 =np.linspace(a, b,100)
    y6 =ag_witch(6,x6)
    
    x7 =np.linspace(a, b,100)
    y7 =ag_witch(7,x7)
    
    ax.set_title(r"$y = \frac{1}{x^2 +1} $   Witch of Agnesi")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_ylim(-1.1, 1.2)

    ax.axhline(y=0.0, color="lightgray")

    ax.plot(x,y,label = r"Exact")
    ax.plot(x3,y3,label = "3 Terms")
    ax.plot(x2,y2, label = "2 Terms")
    ax.plot(x4,y4,label = "4 Terms")
    ax.plot(x5,y5,label = "5 Terms")
    ax.plot(x6,y6,label = "6 Terms")
    ax.plot(x7,y7,label = "7 Terms")
    ax.plot(asymptotex,asymptotey, linestyle="dashed", color="black")
    ax.plot(asymptotex2,asymptotey2, linestyle="dashed", color="black")
    ax.xaxis.set_minor_locator(AutoMinorLocator())
    ax.yaxis.set_minor_locator(AutoMinorLocator())
    

def plot_samples():
    plt.close("all")
    fig = plt.figure(" ")
    gs = fig.add_gridspec(1, 1)

    ax = fig.add_subplot(gs[0, 0])
    plot(ax)
    plt.legend()
    plt.show()
    plt.grid()

plot_samples()
```
